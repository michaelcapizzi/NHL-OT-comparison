```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
first_season = 1990
season_total = 1230
season_2013 = int(float(.59) * season_total)
season_2005 = 0
range_ = range(first_season, 2018)
```

## Regular Season

```python
df_regular_season = pd.DataFrame(columns=["season", "total_games_played", "ot_games_played", "ot_ratio"])
```

```python
for season in range_:
    if season != 2005:
        print("season {}".format(season))
        season_total_final = None
        if season == 2005:
            season_total_final = season_2005
        elif season == 2013:
            season_total_final = season_2013
        else:
            season_total_final = season_total
        data_ = pd.read_html("http://www.hockey-reference.com/leagues/NHL_{}_games.html".format(season))
        data_0 = data_[0]
        data_1 = data_[1]
        data_concat = pd.concat([data_0, data_1])
        series_ = data_concat[data_concat.columns[5]]
        ot = series_[(series_ == "OT") | (series_ == "SO")].count()
        df_regular_season = df_regular_season.append(
            {
                "season": "{}".format(season),
                "total_games_played": "{}".format(season_total_final),
                "ot_games_played": ot,
                "ot_ratio": float(ot) / float(season_total_final) if season_total_final else 0
            },
            ignore_index=True
        )
```

## Playoffs

```python
df_playoff_season = pd.DataFrame(columns=["season", "total_games_played", "ot_games_played", "ot_ratio"])
```

```python
for season in range_:
    if season != 2005:
        print("season {}".format(season))
        data_ = pd.read_html("http://www.hockey-reference.com/leagues/NHL_{}.html".format(season))
        all_games = []
        ot_games = []
        for i in range(3, len(data_)):
            data_indiv = data_[i]
            total_games = data_indiv.shape[0]
            all_games.append(total_games)
            list_ = ["({}OT)".format(i) for i in range(2, 10)]
            ot = data_indiv.isin(list_ + ["(OT)"])
            ot_flattened = ot.values.flatten()
            ot_count = ot_flattened.sum()
            ot_games.append(ot_count)
        all_games_sum = sum(all_games)
        ot_games_sum = sum(ot_games)
        df_playoff_season = df_playoff_season.append(
            {
                "season": "{}".format(season),
                "total_games_played": "{}".format(all_games_sum),
                "ot_games_played": ot_games_sum,
                "ot_ratio": float(ot_games_sum) / float(all_games_sum)
            },
            ignore_index=True
        )
```

```python
df_playoff_season
```

## Visualize

```python
trace1 = go.Bar(
    x=list(range_),
    y=df_regular_season["ot_ratio"],
    name='regular season',
    marker=dict(
        color='red',
    ),
    width=.7,
    opacity=0.6
)
    
trace2 = go.Bar(
    x=list(range_),
    y=df_playoff_season["ot_ratio"],
    name='playoffs',
    marker=dict(
        color='blue',
    ),
    width=.3
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='group',
    title="OT comparison"
)

fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='{}-bar'.format("OT_comparison"))
```
